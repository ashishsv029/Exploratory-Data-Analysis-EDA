```python
#q6 Is there a relation between the length of a movie title and the ratings? 
#    Generate a report that captures the trend of the number of letters in movie titles over years.

lens=[]
for i in dataset['wordsInTitle']:
    #print(i,len(str(i)))
    lens.append(len(str(i)))
dd=pd.DataFrame(lens)
print("The Relation between length of movienames and imdbRatings=",dataset['imdbRating'].corr(dd[0]),'i.e positively correlated')
print()
years=list(dataset['year'])

df2=pd.DataFrame({'year':years,'lettersintitle':lens})
ans=df2.groupby('year')['lettersintitle'].mean()
df3=pd.DataFrame(ans)
df3.reset_index(inplace=True)


def trendline(data,order=1):
    coeffs=np.polyfit(data.index.values,list(data),order)
    slope=coeffs[0]
    return float(slope)

slope=trendline(df3['lettersintitle'])
print("The value of trendline SLOPE is",slope,"which is POSITIVE .Hence the trend is INCREASING")
import matplotlib.pyplot as plt
plt.plot(df3['year'],df3['lettersintitle'],color='red')
plt.xlabel('Year')
plt.ylabel('LettersInMovieTitle')
plt.show()
print("The graph is also showing the increasing trend")
```

```python
#q6 Create a cross tab between the year of the release and the quantile of the movie title length 
#    The results should contain year, min_length, max_length, num_videos_less_than25Percentile, num_videos_25_50Percentile, num_videos_50_75Percentile, num_videos_greaterthan75Precentile

df2=df2.assign(percentiles=pd.qcut(lens, q=[0, .25, 0.50, .75, 1], labels=['<25 percentile', '25-50 percentile', '50-75 percentile','>75  percentile']))
report2=pd.crosstab(df2.year,df2.percentiles)


min_length_df=pd.DataFrame(dataset.groupby('year')['duration'].min())
max_length_df=pd.DataFrame(dataset.groupby('year')['duration'].max())
min_length_df=min_length_df.assign(max_length=max_length_df)
min_length_df.rename(columns={'duration':'min_length'},inplace=True)

report_df=pd.merge(min_length_df,report2,on='year')
report_df.reset_index(inplace=True)
report_df
```

```python
#q7 Generate a report that tracks the average imdb rating quarter on quarter, in the last 10 years, for movies that are top performing. 
#  You can take the top 10% grossing movies every quarter. Add the number of top performing movies under each genre in the report as well.

#last 10 years dataframe
q7df=dataset[(dataset['year']>2010.0) & (dataset['year']<2020.0)]

p=[]
for i ,row in q7df.iterrows():
    if(row['year']>=2011.0 and row['year']<2012.5):
        p.append('q1')
    elif(row['year']>=2012.5 and row['year']<2014.0):
        p.append('q2')
    elif(row['year']>=2014.0 and row['year']<2015.5):
        p.append('q3')
    else:
        p.append('q4')
q7df=q7df.assign(quarters=p)
from collections import Counter
g=q7df.groupby('quarters')

s1=[]
for ii,dff1 in g:
    x={}
    dff1=dff1.sort_values(by=['ratingCount'],ascending=False)
    top10percent=len(dff1)//10
    toppers=dff1.head(top10percent)
    ll1.append(toppers['imdbRating'].mean())
    lss=[]
    for rn,row in toppers.iterrows():
        for jk in row['genrescombo']:
            lss.append(jk)
    x=Counter(lss)
    s1.append(x)

ans3=pd.DataFrame()
d4=[{'quarter':'q1','average':ll1[0]},{'quarter':'q2','average':ll1[2]},{'quarter':'q3','average':ll1[2]},{'quarter':'q4','average':ll1[3]}]
ans3=ans3.append(d4)

emp=pd.DataFrame()
ans2=emp.append(s1)

finalreport1 = pd.concat([ans3, ans2], axis=1)
finalreport1
```

```python
#q8 Bucket the movies into deciles using the duration.
#   Generate the report that tracks various features like nominations, wins, count, top 3 genres in each decile.


time=list(dataset.duration)
dataset2=dataset
dataset2=dataset2.assign(deciles=list(pd.qcut(time,10)))
#dataset2

maxcount=pd.DataFrame(dataset2.groupby('deciles')['ratingCount'].max())
mincount=pd.DataFrame(dataset2.groupby('deciles')['ratingCount'].mean())
maxnominations=pd.DataFrame(dataset2.groupby('deciles')['nrOfNominations'].max())
minnominations=pd.DataFrame(dataset2.groupby('deciles')['nrOfNominations'].sum())
maxwins=pd.DataFrame(dataset2.groupby('deciles')['nrOfWins'].max())
minwins=pd.DataFrame(dataset2.groupby('deciles')['nrOfWins'].sum())
maxrating=pd.DataFrame(dataset2.groupby('deciles')['imdbRating'].max())
minrating=pd.DataFrame(dataset2.groupby('deciles')['imdbRating'].min())


#highest and lowest
finalreport1=pd.merge(maxrating,minrating,on='deciles')
finalreport2=pd.merge(maxcount,mincount,on='deciles')
finalreport3=pd.merge(maxnominations,minnominations,on='deciles')
finalreport4=pd.merge(maxwins,minwins,on='deciles')
ff=pd.merge(finalreport1,finalreport2,on='deciles')
ff1=pd.merge(finalreport3,finalreport4,on='deciles')
finalreport=pd.merge(ff,ff1,on='deciles')

finalreport.rename(columns={'imdbRating_x':'HighRating','imdbRating_y':'lowRating','ratingCount_x':'Highest_Count',
                              'ratingCount_y':'avg_Count','nrOfNominations_x':'MaxNominationsOfAMovie','nrOfNominations_y':'TotalNominations',
                            'nrOfWins_x':'HighestWinsByaMovie','nrOfWins_y':'TotalWins'},inplace=True)

from collections import Counter
g5=dataset2.groupby('deciles')
bglst=[]
for n,dff in g5:
    ls=[]
    for i,row in dff.iterrows():
        for k in row['genrescombo']:
            ls.append(k)
    dct=Counter(ls)
    a,b=[],[]
    for keys in dct:
        b.append(dct[keys])
    b.sort()
    b=b[-1:-4:-1]
    for k1 in b:
        for k2 in dct:
            if(dct[k2]==k1):
                a.append(k2)
    bglst.append(a)
    
finalreport=finalreport.assign(Top3genres=bglst)
finalreport
```
