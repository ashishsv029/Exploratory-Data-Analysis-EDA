# Q1) Sales Dataset

```python
#change the file location accordingly
df=pd.read_excel('C:\\Users\\LRG\\Desktop\\SaleData.xlsx')
df
```

```python
#1Q)Find the least amount sale that was done for each item.
g=df.groupby('Item')
g.min()[['Sale_amt']]
```

```python
#2q)Compute the total sales for each year and region across all items
g1=df.groupby([df['OrderDate'].dt.year,'Region'])
q2=g1.sum()[['Sale_amt']]
q2
```

```python
#Q3)Create new column 'days_diff' with number of days difference between a reference date passed and each order date
reference_date=pd.to_datetime('2020/01/01')
df=df.assign(days_diff=reference_date-df['OrderDate'])
df
```

```python
#4Q)Create a dataframe with two columns: 'manager', 'list_of_salesmen'. Column 'manager' will contain the unique managers present and column 'list_of_salesmen' will contain an array of all salesmen under each manager.
import numpy as np
grp4q=df.groupby('Manager')
managers_list,salesman_list=[],[]
for manager,manager_df in grp4q:
    salesmanarray=np.array(list(set(manager_df['SalesMan'])))
    managers_list.append(manager)
    salesman_list.append(salesmanarray)

df4q=pd.DataFrame(list(zip(managers_list,salesman_list)),columns=['Manager','list_of_salesman'])
df4q
```

```python
#5Q)For all regions find number of salesman and total sales. Return as a dataframe with three columns - Region, salesmen_count and total_sales
g4=df.groupby('Region')
q5=pd.DataFrame(g4.count()[['SalesMan']])
q3=pd.DataFrame(g4.sum()[['Sale_amt']])
q4=pd.DataFrame(q3)
df_new=pd.concat([q5,q4],axis=1)
df_new=df_new.rename(columns={'SalesMan':'salesman_count','Sale_amt':'total_sales'})
df_new
```

```python
#6Q)Create a dataframe with total sales as percentage for each manager. Dataframe to contain manager and percent_sales
total_sales=df['Sale_amt'].sum()
grp=df.groupby('Manager')
df6=grp.sum()[['Sale_amt']]
q6df=df6[['Sale_amt']]/total_sales*100
q6df.columns
q6df.rename(columns={'Sale_amt':'percent_sales('+'%)'},inplace=True)
q6df
```

# Q2) DIAMOND DATASET

```python
#ENTER THE FILE LOCATION OF YOUR SYSTEM
df2=pd.read_csv("C://Users//ASHISH//Desktop//Diamonds.csv",na_values=['None'])
```

```python
#q1)Count the duplicate rows of diamonds dataframe.
df2.drop_duplicates(inplace=True)
rows,cols=df2.shape
print("No of unique rows after deleting duplicate rows=",rows)
df2
```

```python
#2Q)Drop rows in case of missing values in carat and cut columns.
df2 = df2.dropna(how='all',subset=['carat', 'cut'])
df2.shape
```

```python
#3q)Subset the dataframe with only numeric columns
dfnum = df2.select_dtypes(include=np.number)
dfnum
```

```python
#4Q)Compute volume as (x*y*z) when depth is greater than 60. In case of depth less than 60 default volume to 8.
df2['z'].fillna(df2['z'].mean(),inplace=True)
vol=[]
for i,row in df2.iterrows():
    if row['depth']>60:
        vol.append(row['x']*row['y']*row['z'])
    else:
        vol.append(8)
df2=df2.assign(volume=vol)
df2
```

```python
#5Q)Impute missing price values with mean.
df2['price'].fillna(df2['price'].mean(),inplace=True)
df2
```

```python
#6q)In diamonds data set Using the volume calculated above, create bins that have equal population within them. Generate a report that contains cross tab between bins and cut. Represent the number under each cell as a percentage of total.
df3=df2.assign(range=list(pd.qcut(df2.volume, 10)))
pd.crosstab(df3.cut,df3.range,normalize='index')
```
